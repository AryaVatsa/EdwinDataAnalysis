```python
listOfDates = []
SymptomsList = []
otherSymptoms = []
for n in range(8):
    SymptomsList.append("yes")
for n in range(2):
    SymptomsList.append("no")

for n in range(1):
    otherSymptoms.append("yes")
for n in range(19):
    otherSymptoms.append("no")

for n in range(8, 15):
    listOfDates.append(str(n) + "0124")
for i, row in enumerate(data["Did you experience any symptoms?"]):
    symptomsYesNo = random.choice(SymptomsList)
    data.at[i, "Did you experience any symptoms?"] = symptomsYesNo
    if symptomsYesNo == "yes":
        data.at[i, 'Did you experience fever?'] = random.choice(["yes", "no"])
        data.at[i, 'What is the earliest onset date of your symptoms'] = random.choice(listOfDates)
        data.at[i, "What is the earliest onset time of your symptoms"] = random.randrange(100, 2300, 100)
        data.at[i,"What was the highest recorded temperature"] = (random.randrange(375, 400))/10
        VomitYesNo = random.choice("yes", "no")
        data.at[i,"Did you experience vomiting?"] = VomitYesNo
        if VomitYesNo == "yes":
            data.at[i,"How many times did you vomit?"] = random.randrange(1, 9, 1)

        diarrhoeaYesNo = random.choice("yes", "no")
        data.at[i, "Did you experience diarrhoea?"] = diarrhoeaYesNo
        if diarrhoeaYesNo == "yes":
            data.at[i,"Please indicate the most number of episodes that you have experience diarrhoea in 24 hours"] = random.randrange(1, 9, 1)
            data.at[i,"descripe the colour of your loose stools"] = random.choice(["black", "Blue", "Brown", "Clear watery", "Green", "Grey", "Red", "White / Pale"])
        
        data.at[i,"Did you experience nausea?"] = random.choice("yes", "no")
        data.at[i,"Did you experience abdominal pain?"] = random.choice("yes", "no")
        data.at[i,"Did you experience headache?"] = random.choice("yes", "no")
        data.at[i,"Other symptoms"] = random.choice(otherSymptoms)
        data.at[i, "Were you unwell with GE symptoms (diarrhoea/vomiting) in the past 2 weeks?"] = random.choice("yes", "no")
        data.at[i,"What is the earliest date you had diarrhoea / vomitting in the past 2 weeks"] = random.randrange(1,8)
        householdYesNo = random.choice(['yes', 'no'])
        data.at[i,"Has any of your household members experience GE illness within the week?"] = random.choice(['yes', 'no'])
        if householdYesNo == "yes":
            data.at[i,"What is the date of the symptom onset for your household members?"] = random.choice(listOfDates)
            data.at[i,"Please indicate the number of household members that experienced diarrhoea/vomiting after your symptom onset"] = random.randrange(1,6)
        
        householdYesNo = random.choice(["yes", "no"])
        data.at[i,"Do you have household members who were unwell with diarrhoea/vomiting symptoms in the past 2 weeks before your symptoms onset?"] = householdYesNo
        friendDiarOrVomit = random.choice(["yes", "no"])
        data.at[i,"Do you have any friends/colleagues who were unwell with diarrhoea/vomiting symptoms in the past 2 weeks before your symptoms onset?"] = friendDiarOrVomit
        if friendDiarOrVomit == "yes":    
            data.at[i,"When did household members / friends / colleagues experience diarrhoea/ vomiting in the past 2 weeks?"] = random.choice(listOfDates)
for i, row in enumerate(data['[MyInfo] Name']):
    data.at[i, '[MyInfo] Name'] = ''.join([random.choice(string.ascii_letters) for u in range(10)])
    data.at[i, 'NRIC/FIN'] = "t" + str(i) + "x"
    data.at[i, 'sgID Validated NRIC'] = "t" + str(i) + "x"
    data.at[i, '[MyInfo] Gender'] = random.choice(["m", "f"])
    data.at[i, '[MyInfo] Date of birth'] = str(random.randrange(1, 30)) + str(random.randrange(1, 13)) + str(random.randrange(1999, 2005))
    data.at[i, "[MyInfo] Race"] = random.choice("chinese", "malay", "indian", "others")
    data.at[i, "Mobile number"] = random.randrange(80000000, 99999999)
    data.at[i, "Which camp are you from?"] = random.choice(CampOptions)
    data.at[i, "Which company (COY) are you from?"] = random.randrange(1,4)
    data.at[i, "Which Platoon are you from?"] = random.randrange(1,5)
    data.at[i, "Which section are you from?"] = random.randrange(1,5)
    data.at[i, "What is your bed no?"] = random.randrange(1,17)
    data.at[i, 'Date of Outbreak declared'] = random.choice("150124", "160124", "170124")

data
```
